## 数据预处理

**(1) 数据拆分**

我们在每一类鸢尾中都抽取70%的数据用于训练，剩下30%的数据用于测试：

```python
# sklearn中的train_test_split函数用于数据拆分
from sklearn.model_selection import train_test_split
# 将数据iris带入train_test_split函数，test_size=0.3意思是从中抽取30%的数据用于测试数据。
# train_test_split函数的返回值是两个数据表，第一个包含70%的数据（存为train，意为训练数据），第二个包含30%的数据（存为test，意为测试数据）。
train, test = train_test_split(iris, test_size = 0.3)
# 将训练数据train的predictor和outcome拆分。
train_X = train[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]# taking the training data features
train_y = train['Species']
# 将测试数据test的predictor和outcome拆分。
test_X= test[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']] # taking test data features
test_y =test['Species']
```

**(2) 数据标准化**

从前面的分析（数据总结表&散点图）中，我们可以看到横纵坐标轴的范围不一样，因为大花瓣和小花瓣的维度不一样。具体数据如下：

* 大花瓣的长度在4.3到7.9厘米之间
* 大花瓣的宽度在2.0到4.4厘米之间
* 小花瓣的长度在1.0到6.9厘米之间
* 小花瓣的宽度在0.1到2.5厘米之间

基于KNN计算两个observation的相似性的方法需要计算每一个predictor值之间的差异，所以如果两个predictor的取值范围不一样，那么他们对“相似度”的影响就不一样。通常，我们会平等的对待每一个predictor，所以我们需要将它们的范围划定到同样的区间。这一步叫做数据标准化（Normalization）。接下来，我们就将每一个predictor的数据维度“缩放”为平均值为0，标准差为1的分布。

注意，为了保证训练数据和测试数据之间没有关联，我们需要先拆分数据，再分别正规化数据。感兴趣，希望深入了解的同学可以参考Stackoverflow的这个问答：[Normalize data before or after split of training and testing data?](https://stackoverflow.com/questions/49444262/normalize-data-before-or-after-split-of-training-and-testing-data)

```python
# 从sklearn中引入正规化数据的工具StandardScaler
from sklearn.preprocessing import StandardScaler
# 初始化一个StandardScaler()，命名为sc
sc = StandardScaler()
# 将训练数据灌入
sc.fit(train_X)
train_X_std = sc.transform(train_X)
test_X_std = sc.transform(test_X)
```

## 建立模型

**(1) 第一个KNN模型**

我们会使用sklearn来建立KNN模型。首先我们随机选取一个K，令K等于3来建立模型。

```python
# 引入sklearn的KNN类
from sklearn.neighbors import KNeighborsClassifier
# 引入用于评估模型准确度的metrics方法
from sklearn import metrics
# 将K值3赋予模型参数n_neighbors，初始化knn模型
knn = KNeighborsClassifier(n_neighbors=3)
# 将训练数据灌入knn模型的fit函数进行模型训练，完成这一步之后knn就是我们训练好的模型
knn.fit(train_X_std, train_y)
# 使用knn的predict函数对预测数据test_X_std进行处理，得到预测值，存储为pred_y
pred_y = knn.predict(test_X_std)
# 利用metrics评估预测值相对outcome的准确度
print('当K=3时，KNN模型的准确度是： ' + str(metrics.accuracy_score(test_y, pred_y)))
```

**(2) K的选择**

以上，我们得到了当K等于3时，模型的准确度。但有没有其他的K值，让KNN模型表现的更好呢？

为了回答这个问题，我们会让K取不同的值，然后分别训练模型进行比较。通常，我们给K设定的范围在1-50之间。现在我们测试一下K在这个范围内，模型的表现。

```python
# k_range是一个包含1到50之间整数的数组，注意51不在数组中。
k_range = list(range(1,51))
# 初始化score这个数组，50个k值对应的50个模型的准确度都会存储在score中。
scores = []
# 我们用for循环将每一个k值带入KNN模型进行训练，然后把结果存储在score数组中。
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(train_X_std, train_y)
    pred_y = knn.predict(test_X_std)
    scores.append(metrics.accuracy_score(test_y, pred_y))
# 将结果绘制成折线 
plt.plot(k_range, scores)
# 增加X轴的标签，X轴代表不同k值
plt.xlabel('Value of k for KNN')
# 增加Y轴的标签，Y轴代表模型的准确度
plt.ylabel('Accuracy Score')
# 增加标题
plt.title('Accuracy Scores for Values of k of k-Nearest-Neighbors')
# 将图片显示出来
plt.show()
```

我们可以看到，折线图整体是先上升再下降的趋势。我们在上一章教程中讲解的是错误率（Error Rate），所以对应曲线是凹下去的。这里Y轴表达的是准确率（Accuracy Score），所以图形是凸起来的。

但是我们能够看到很多的锯齿，并不是平滑的连续变化，这说明预测结果结果具备一定的随机性。造成这个随机性的原因是：

**数据量太小（150个observation），所以拆分测试数据和训练数据的随机性会对结果带来较大的影响。**

所以，我们很难通过这个图选择一个合适的K值。理想情况是我们会看到一条平滑变化的曲线，然后选取曲线中最高点对应的K值。接下来，我们就来解决这个问题，使得K值选取更准确。

**(3) 多次拆分数据，训练模型**

由于随机性具有无方向性，如果将多次随机的结果加和平均，随机性造成的误差就会互相抵消，我们就会得到相对准确的结果。

借助这个思路，我们可以将数据拆分，正规化，以及模型训练重复n次。这样对于每个K值，我们就可以得到n个模型，以及对应的n个准确度（Accuracy Score）。将每个K值准确度取平均，理论上就能够得到更准确的准确度，画出一条更平缓的准确度/K值曲线，从而可以更准确的选择K值。

接下来，我们就把上面的数据拆分，正规化，模型训练进行1000遍。然后取平均结果，来评估模型的表现和K值的选取。这段代码大约会运行5分钟。

```python
# 本质上，我们要做的事情是在上段代码的外层叠加一个for循环，并且加入数据拆分的代码。让数据拆分和模型训练一起运行1000次

# result数组中会存储1000个scores数组。（scores的定义可以参考上一段代码的注释）
result = []
# 让内部代码运行1000次的for循环
for i in range(1000):
    # 拆分数据
    train, test = train_test_split(iris, test_size = 0.3)
    train_X = train[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]# taking the training data features
    train_y = train.Species# output of our training data
    test_X= test[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']] # taking test data features
    test_y =test.Species   #output value of test data
    # 正规化数据
    sc = StandardScaler()
    sc.fit(train_X)
    train_X_std = sc.transform(train_X)
    test_X_std = sc.transform(test_X)
    # 训练模型
    k_range = list(range(1,50))
    scores = []
    for k in k_range:
        knn = KNeighborsClassifier(n_neighbors=k)
        knn.fit(train_X_std, train_y)
        pred_y = knn.predict(test_X_std)
        scores.append(metrics.accuracy_score(test_y, pred_y))
    result.append(scores)

# 将每一个K值对应模型的准确率（Accuracy Score）取平均
result = np.array(result)
result = np.mean(result, axis=0)
# 绘图
plt.plot(k_range, result)
plt.xlabel('Value of k for KNN')
plt.ylabel('Accuracy Score')
plt.title('Accuracy Scores for Values of k of k-Nearest-Neighbors')
plt.show()
```

```python
# 将结果打印出来
result
```

我们可以看到，当K等于11的时候，模型的准确率最高，超过95.5%。所以基于这个鸢尾数据库的KNN模型，我们会选择K值为11的KNN模型。

**(4) 预测**

接下来，我们就用K=11对所有数据训练训练KNN模型，同时预测下面这个新的observation的种类：

* 大花瓣的长度：6
* 大花瓣的宽度：3
* 小花瓣的长度：4
* 小花瓣的宽度：2

```python
# 从iris删除Id，Species两列值，得到包含所有observation的predictor的数据X
X = iris.drop(['Id', 'Species'], axis=1)
# 取出outcome
Y = iris['Species']
# 初始化KNN模型
knn = KNeighborsClassifier(n_neighbors=11)
# 将X和Y灌入模型进行训练
knn.fit(X, Y)
# 将新的observation的predictor灌入进行预测
knn.predict([[6, 3, 4, 2]])
```

所以，这个新的鸢尾植物的种类就是Iris-versicolor。长这样：
