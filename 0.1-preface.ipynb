# Data Science for Everyone

### 为什么要学习Data Science
一句话解释清楚：
> 做出更“优”的决策

我们每天都在有意识或者无意识的做出诸多的决策，小到吃什么早餐，大到如何投资。

这些决策的大部分依赖我们的直觉，而且在大多数情况下这是非常高效的做法。但是，我们也时常被直觉欺骗...

我们之所以能够快速选择正确的食物，是因为在进化和成长过程中，我们在持续训练自己“如何觅食”、吃哪一种食物能够让我们获得需要的营养。我们大脑经过了漫长的自然选择，加上后天经年累月的学习，让我们能够迅速且正确地做出“吃什么”的决策。

但是，在某些领域我们缺乏这些训练，导致我们的直觉并不适用于这些领域。举一个简单的例子，证券交易。证券市场起源自1602年，最先由一个叫Dutch East India Co.的公司发行。400年的时间，相对于人类进化历史来讲如同一瞬间，所以对于股票的决策并没有刻画在我们的基因里。同时，大部分人也没有学习过相关的交易知识。
> 但我们盲目地利用直觉做买卖股票的决策，并且迷之自信。

作为资深韭菜，我有深刻的体会。2014年那时候我还在百度工作，朋友给我讲了个“内部消息”并基于此强烈推荐我买一只股票。我一听觉得自己仿佛要“财务自由”了，于是将一年攒下的积蓄都投入到一只股票里面。结果是，三个月亏了近50%。事后反思如下：
> * 我做决策的逻辑是基于对朋友的“信任”
> * 朋友出于好心告知我消息，他完全相信自己的判断
> * 朋友没有经历过股票投资的训练，“内部消息”的真实性和实效性也并不清楚，他做出正确判断的概率并不高

我被直觉蒙蔽了，认为“内部消息”等同于准确判断，认为“信任的人”说出来的话都是可信度高的。这些逻辑错误导致我自信的孤注一掷，也导致了50%的亏损。这同时是为什么大部分人会亏钱的原因。

对于股票市场，使用Data Science来进行量化分析是一个比使用“直觉”更好的一个切入点。有很多Data Scientist利用自己的专业知识在股票市场稳定的挣到了钱，有更多的Data Scientist利用专业知识提前判断出他们无法在股票市场稳定的挣到钱，从而避免了潜在的损失。

当然，股票市场只是一个例子。Data Science的应用范围，随着这个“数字化”时代的发展成指数倍增加的。这也是要学习Data Science的另一个原因。

### 永恒的不确定性
你可能听过这句话：这个时代唯一不变的就是变化。这句话的Data Science版本是：
> 这个世界唯一确定的就是不确定性

很多人对于Data Science的一个误解是：以为Data Science能够告诉我们一个准确的结果，一个准确的选项。但这是应试教育的思维方式，并不符合真实的世界模型。实际上，发生在真实世界里的事情，很少有准确的答案。

Galton board这个弹珠游戏大家应该并不陌生。弹珠从顶部被放下，经过一系列的左右移动，最终会落在底部的某一个格子里。

每当我们放下一颗弹珠，无论我们有多么想，都无法确定的知道弹珠将会落在哪一个格子里。我们能做的，是预估出弹珠下落在每个格子里的概率。如果真的要做出一个预测，那么我们就选择概率最大的那一个格子。虽然我们无法保证预测是对的，但是我们可以做出胜率最高的预测，并且准确衡量预测对的概率是多少。

综上所述，Data Science真正的能力是：
> * 帮助我们尽可能准确的描绘不确定性(概率分布 Distribution)
> * 然后我们再利用这个“准确的不确定性”来做决策。

### 反直觉的决策

思考一下这个问题：
> 有三个外表相同的盒子，其中一个里面有张一百万的支票。<br>
> 首先，你需要从中选择一个盒子。然后我会打开另外两个盒子其中的一个空盒子。<br>
> 然后，我给你一个选项，要不要改选另一个盒子。

你可以用两分钟思考一下这个问题。

第一次面对这个问题，我的直觉是：剩下两个盒子的中奖概率是一样的。

但实际上通过概率计算我们可以得出，最优策略是改选另一个盒子，获奖概率会从1/3增加到2/3。

具体的计算过程也很简单。一开始，奖品在选中的盒子中的概率为1/3，在另外两个盒子中的概率为2/3。在我排除另外两个盒子中的一个之后，2/3的概率分布并没有变化，只是都被赋予给了这两个盒子中的另一个。

通过这个例子，也许你能够有这样一种感受：
> Data Science帮助我们做出**“反直觉但正确”**的选择
